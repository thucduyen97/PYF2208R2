```python
#vong lap while

import turtle
t=turtle.Turtle()
t.speed(50)

khoang_cach=int(input('Nhap khoang cach: '))
count=1
r=2

while count<khoang_cach:
    r+=5
    t.circle(r,180)
    count+=1
    
turtle.done()
```

```python
#vong lap for

import turtle
t=turtle.Turtle()
t.speed(50)

khoang_cach=int(input('Nhap khoang cach: '))

for x in range(khoang_cach):
    t.circle(x,-30)
turtle.done()
```
