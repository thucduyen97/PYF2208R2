## Viet chuong trinh nhap vao diem trung binh cua 3 ban a,b,c => ban nao co diem trung binh lon nhat

```python
per1=int(input('Nhập điểm trung bình học sinh a: '))
per2=int(input('Nhập điểm trung bình học sinh b: '))
per3=int(input('Nhập điểm trung bình học sinh c: '))

if per1>per2 and per1>per3:
    print(f'Điểm trung bình lớn nhất là {per1}')
else:
    if per2>per3:
        print(f'Điểm trung bình lớn nhất là {per2}')
    else:
        print(f'Điểm trung bình lớn nhất là {per3}')
```

## Cấu trúc lặp while

### Ví dụ: In ra tất cả số chẵn từ 0 đến n



```python
n=int(input('Input n: '))
count=0
while count<=n:
    print(count)
    count+=2
```

### Ví dụ: In ra tất cả số lẻ từ n về 1

```python
n=int(input('Input n: '))
while n>=1:
    if n%2==1:
        print(n)
    n-=1
```

### Ví dụ: Đảo ngược chuỗi

## Input 1 chuoi eg: abcd => dcba

```python
n=(input('Input string: '))
i=len(n)-1 #kiem tra ki tu cuoi cung co index la bao nhieu

while i>=0:
    print(n[i],end='')
    i-=1
```

### Ví dụ: Vẽ đa giác đều

### Ví dụ: 
Nhập vào một số nguyên a

Cho S(k) = 1 + 2 + 3 + ... + k

Lựa chọn k sao cho S(k) > a với k nhỏ nhất

In S(k) và k lên màn hình

```python
a=int(input('Input a: '))
S=0
k=1

while S<=a:
    k+=1
    S+=k
    print(k,S)
```

### Lệnh Break

Cú pháp:

`break`

Lệnh break sẽ giúp hủy vòng lặp đang thực hiện

##### Lưu ý:

##### - Lệnh break <b>BẮT BUỘC</b> phải được đặt trong vòng lặp
##### - Lệnh break sẽ hủy vòng lặp gần nhất (trong trường hợp có nhiều vòng lắp lồng vào nhau)

#### Ví dụ

Sử dụng lệnh `while (True)` và break và in ra 10 lần câu "Hello mọi người"

```python
while True:
    print('Hello World')
    break
```

## Lệnh continue

## In ra nhung so chia het cho 4 trong khoang tu 0 den 20, tru so 8

```python
count=0
while count<=20:
    count+=4
    if count ==8:
        continue
    print(count)
```
