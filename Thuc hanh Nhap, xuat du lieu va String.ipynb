## Bài 3
Nhập vào 'tien_luong_theo_gio' và 'so_gio_lam_viec_trong_thang' và 'tien_thuong'
In ra Tổng thu nhập = tiền lương theo giờ*số giờ+tiền thưởng


```python
a=float(input("tien_luong_theo_gio: "))
b=float(input("so_gio_lam_viec_trong_thang: "))
c=float(input("tien_thuong: "))
print('Tong thu nhap',a*b+c)
```

## Bài 5
Nhập 1 'string' kiểm tra xem đây có phải là 1 'string' đối xứng không
Note: chuỗi đối xứng là chuỗi có thứ tự giống nhau khi đọc từ trái sang phải và ngược lại.

eg: 
- abba
- abcddcba

```python
string=(input('Nhập string: '))
stringnguoc=string[::-1]
print('Đối xứng: ',string==stringnguoc)
```
