# [Bài tập] Vẽ nhà sử dụng hàm

```python
import turtle

t = turtle.Turtle()
def draw(width = 100, height = 200): 
    t.speed(0)
    t.penup()
    t.goto(-100,0)
    t.pendown()

    for i in range(2):
        t.fd(width)
        t.rt(90)
        t.fd(height)
        t.rt(90)
    t.fd(height)
    for i in range(2):
        t.rt(90)
        t.fd(height)
        t.rt(90)
        t.fd(width)
    for i in range(2):
        t.fd(width+50)
        t.rt(90)
        t.fd(height)
        t.rt(90)
        
    t.penup() #di den ve cua chinh
    t.goto(150,-200)
    t.pendown()
    
    t.begin_fill() #vecua
    t.fillcolor("purple")
    t.lt(90)
    t.fd(height-150)
    t.rt(90)
    t.fd(width-50)
    t.rt(90)
    t.fd(height-150)
    t.end_fill()

    t.penup() #di den ve cua so
    t.goto(20,-150)
    t.pendown()
    
    t.begin_fill() #vecuaso
    t.fillcolor("purple")
    for i in range(2):
        t.lt(90)
        t.fd(width-70)
        t.lt(90)
        t.fd(height-130)
    t.end_fill()
    
    t.penup() #di den ve cua so
    t.goto(-80,-150)
    t.pendown()
    
    t.begin_fill() #vecuaso
    t.fillcolor("purple")
    for i in range(2):
        t.lt(90)
        t.fd(width-70)
        t.lt(90)
        t.fd(height-130)
    t.end_fill()
    
    t.penup() #di den ve mai nha
    t.goto(100,0)
    t.pendown()
    
    t.begin_fill() #vecuaso
    t.fillcolor("yellow")
    t.rt(180)
    t.rt(50)
    t.fd(width)
    t.rt(80)
    t.fd(width)
    t.end_fill()
    
    t.penup() #di den ve mai nha
    t.goto(-100,0)
    t.pendown()
    
    t.begin_fill() #vecuaso
    t.fillcolor("orange")
    t.lt(100)
    t.fd(width-25)
    t.rt(60)
    t.fd(240)
    t.rt(140)
    t.fd(width+2)
    t.end_fill() 
    
    turtle.done()
    

draw(100,200) 
```
