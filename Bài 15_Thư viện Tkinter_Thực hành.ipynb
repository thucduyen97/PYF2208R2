```python
#Duyen

from tkinter import *
from tkinter import messagebox

window = Tk()
window.title('My intro')

title_lbl = Label(window, text = 'My introduction')
firstname_lbl = Label(window, text = 'First name')
lastname_lbl = Label(window, text = 'Last name')

firstname = Entry(window)
lastname = Entry(window)

def get_info():
    messagebox.showinfo('User info',f'My name is: {firstname.get()} {lastname.get()}')


submit_btn = Button(window, text ='Submit', command = get_info)

title_lbl.grid(row = 0, column = 0, columnspan = 2)
firstname_lbl.grid(row = 1, column = 0)
lastname_lbl.grid(row = 2, column = 0)
firstname.grid(row = 1, column = 1)
lastname.grid(row = 2, column = 1)
submit_btn.grid(row = 4, column = 0)

window.mainloop()
```

# Bài 2: Ứng dụng tính BMI

```python
#Duyen

from tkinter import *
from tkinter import messagebox

window = Tk()
window.title('BMI')

bmi_lbl = Label(window, text = 'BMI Calculation')
weight_lbl = Label(window, text = 'Your weight (kg)')
height_lbl = Label(window, text = 'Your height (m)')

weight = Entry(window)
height = Entry(window)

def get_info():
    result = 'Your BMI is:' + ' ' + str(float(weight.get()) / (float(height.get()) * 2))
    type_check = float(weight.get()) / (float(height.get()) * 2)
    
    if type_check < 16:
        messagebox.showinfo('User info',f'{result} Gay cap do III')
    elif type_check < 17:
        messagebox.showinfo('User info',f'{result} Gay cap do II')
    elif type_check < 18.5:
        messagebox.showinfo('User info',f'{result} Gay cap do I')
    elif type_check < 25:
        messagebox.showinfo('User info',f'{result} Binh thuong')
    elif type_check < 30:
        messagebox.showinfo('User info',f'{result} Thua can')
    elif type_check < 35:
        messagebox.showinfo('User info',f'{result} Beo phi cap do I')
    elif type_check < 40:
        messagebox.showinfo('User info',f'{result} Beo phi cap do II')
    else:
        messagebox.showinfo('User info',f'{result} Beo phi cap do III')


submit_btn = Button(window, text ='Submit', command = get_info)

bmi_lbl.grid(row = 0, column = 0, columnspan = 2)
weight_lbl.grid(row = 1, column = 0)
height_lbl.grid(row = 2, column = 0)
weight.grid(row = 1, column = 1)
height.grid(row = 2, column = 1)
submit_btn.grid(row = 4, column = 0)

window.mainloop()
```
