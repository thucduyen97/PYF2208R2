# 1B. Quản lý hình học

Phương thức hỗ trợ quản lý, sắp xếp các tiện ích trên giao diện Tkinter.  
Để setup độ rộng cửa sổ giao diện.



VD

- from tkinter import *
- window = Tk()
- window.title("Practice - Tkinter Module")
- window["bg"] = "Black"
- window.geometry("430x250")
- window.mainloop()

Có 3 phương thức quản lý chính: grid, pack, place

* Phương thức 1: grid  
    - Đây là phương thức phổ biến nhất. Theo cấu trúc table -lưới, sử dụng hàng và cột để sắp xếp các tiện ích.
    - Giao điểm của hàng và cột là 1 ô, chỉ số mặc định của ô đầu tiên là (cột, hàng) = (0,0). 
    - Mỗi tiện ích được sắp sếp vào 1 ô và mỗi ô chỉ chứa 1 tiện ích.


syntax: widget.grid(option)



* Các option của grid




```python
# Ví dụ với row, column, columnspan
from tkinter import *
window = Tk()
window.title("Practice - Tkinter Module ")
window.geometry("430x250")
lb1 = Label(window, text = 'Nội dung 1')
lb1.grid(row = 0, column = 0)

lb2 = Label(window, text = 'Nội dung 2', font = ('Arial', 12, 'bold'))
lb2.grid(row = 1, column = 0)

lb3 = Label(window, text = 'Nội dung 3', font = ('Arial', 15, 'bold'), fg = 'red', bg = 'black')
lb3.grid(row = 0, column = 1, sticky = W)

lb4 = Label(window, text = 'Nội dung 4', font = ('Arial', 20, 'bold'), fg = 'red', bg = 'yellow')
lb4.grid(row = 1, column = 1)

lb5 = Label(window, text = 'Nội dung 5', font = ('Arial', 20, 'bold'), fg = 'red', bg = 'aqua')
lb5.grid(row = 2)

window.mainloop()
```

```python
#Ví dụ với padx, pady
from tkinter import *
window = Tk()
window.title("Practice - Tkinter Module ")
window.geometry("430x250")
lb2 = Label(window, text = 'Nội dung 2', font = ('Arial', 12, 'bold'))
lb2.grid(row = 1, column = 0, padx = (20,0), pady = 30)

lb5 = Label(window, text = 'Nội dung 5', font = ('Arial', 12, 'bold'), fg = 'red', bg = 'aqua')
lb5.grid(row = 1, column = 1, padx = 20, pady = 30)
window.mainloop()
```

```python
### NHÁP!!! KHÔNG NẰM TRONG PHẦN TRÌNH BÀY

from tkinter import *
a = Tk()
a.title('Registration')
a.geometry('270x300') #kích thước box rộng x cao
a['bg'] = 'grey' #màu background
a.attributes('-topmost', True) #box luôn luôn hiện lên trên các tab khác

#tạo text box
name = Label(a, text = 'Please click below for registering!', font = ('Arial', 14), bg = 'red', fg = 'white')
name.place(x = 80, y = 100) #từ vị trí đầu tiên góc phải, x đi ngang, y đi xuống

#taọ button:
# bg: background color
#fg: foreground color (text color)
def Clickresult():
    name1 = Label(a, text = 'Thank you for registering!', font = ('Arial', 14), bg = 'red', fg = 'white')
    name1.place(x = 90, y = 250)
    #có thể thêm nhiều câu lệnh popup cùng lúc
    
but = Button(a, text = 'Click here', width = 15, height = 5, bg = 'yellow', font = ('Times New Roman', 13), fg = 'red', command = Clickresult)
# command là câu lệnh sẽ hiển thị sau khi click vào button, 
# vì thế phải thêm định nghĩa hàm Clickresult vào, nếu không sẽ báo lỗi
but.place(x = 110, y = 130)

 
a.mainloop()
```

```python
# 2. textvariable

# textvariable được sử dụng để cung cấp giá trị thông qua các biến
# kiểu dữ liệu: Integer or String
# integer : IntVar() 
# String: StringVar() 

from tkinter import * 

a = Tk()
a.title("Your question")
a.geometry("250x200")
name = StringVar(a, value='Hello world')
nameTf = Entry(a, textvariable=name).pack()

a.mainloop()
```

```python
from tkinter import *
from tkinter import messagebox

ws = Tk()
ws.title('pythonguides')
ws.geometry('250x200')

def welMsg(name):
    name = name_Tf.get()
    return messagebox.showinfo('pythonguides', f'Hi! {name}, welcome to pythonguides' )

Label(ws, text='Enter Name & hit Enter Key').pack(pady=20)
name_Tf = Entry(ws)
name_Tf.bind('<Return>',welMsg)
name_Tf.pack()

ws.mainloop()
```

```python
#Liên kết label, entry, button
#Thực hành liên kết label, entry, Button
#Tạo ra cửa số chương trình
from tkinter import *
learn = Tk ()
learn.title ('tim hieu ve Tkinter')
learn.geometry ('300x300')
learn.attributes ('-topmost', True)


#Tạo ra Label
name = Label(learn, font = ('Arial',12), text = 'Họ và tên: ')
name.place(x = 10, y = 10)

#Tạo ra Entry
entry = Entry(learn, width = 30, font = ('Times New Roman',10))
entry.place (x=80, y=10)
entry.focus ()

#Tạo ra Button
def clickresult():
    name1 = Label(learn, font = ('Arial',10), text = 'Bạn đã nhập:' + entry.get())                                                                                             
    name1.place(x = 50, y = 70)
but = Button(learn, text = 'Click vao day', width = 10, height = 1, bg = 'yellow', font = ('Times New Roman',13),  
             command = clickresult)
but.place (x = 80, y = 30)

learn.mainloop ()
```

```python
## Bài toán cộng hai số thực
from tkinter import *
learn = Tk ()
learn.title ('tim hieu ve Tkinter')
learn.geometry ('300x300')
learn.attributes ('-topmost', True)


#Tạo ra Label1
name1 = Label(learn, font = ('Arial',12), text = 'Số thứ 1: ')
name1.place(x = 10, y = 10)

#Tạo ra Label2
name2 = Label(learn, font = ('Arial',12), text = 'Số thứ 2: ')
name2.place(x = 10, y = 50)


#Tạo ra Entry1
entry1 = Entry(learn, width = 30, font = ('Times New Roman',10))
entry1.place (x=80, y=10)
entry1.focus ()

#Tạo ra Entry2
entry2 = Entry(learn, width = 30, font = ('Times New Roman',10))
entry2.place (x=80, y=50)

#Tạo ra Button
def clickresult ():
    name1 = Label(learn, font = ('Arial',10), text = 'Kết quả phép tính là:'+ str(float(entry1.get()) + float(entry2.get())))
    name1.place(x = 50, y = 70)
but = Button(learn, text = 'Cộng 2 số', width = 10, height = 1, bg = 'yellow', font = ('Times New Roman',13),  
             command = clickresult)
but.place (x = 80, y = 100)
    
    
    
learn.mainloop ()
```
