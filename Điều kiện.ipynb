## Câu lệnh if..else

```python
mark = float(input("Nhập điểm cuối kỳ: "))
if mark >= 8.5:
    print("Bạn được tuyển thẳng")
else:
    print("Bạn chưa đủ điểm")
```

### Thực hành:
Viết chương trình nhập vào một số nguyên và cho biết số nguyên đó là số chẵn hay số lẻ.


```python
num = int(input("Nhập số nguyên: "))
if num % 2 == 0:
    print(f"{num} là số chẵn")
else:
    print(f"{num} là số lẻ")
```

### Thực hành:
Viết chương trình nhập vào một chuỗi và cho biết trong chuỗi đó có chứa cụm từ "python" hay không.


```python
string = input("Nhập chuỗi: ")
if "python" in string:
    print(f"'{string}' có chứa cụm từ 'python'")
else:
    print(f"'{string}' không chứa cụm từ 'python'")
```

## Câu lệnh if lồng nhau

```python
# Nhập vào tổng số giờ làm và level của nhân viên. Nhân viên sẽ được khen thưởng nếu như tổng số giờ làm > 800 giờ
# Mức độ khen thưởng như sau
# - Nếu là nhân viên level 1 thì thưởng 1000
# - Nếu là nhân viên level 2 thì thưởng 2000

total_hour=int(input('Nhập tổng số giờ làm: '))
level=int(input('Nhập level nhân viên: '))

if total_hour > 800:
    if level==1:
        print('Bạn được nhận thưởng 1000USD')
    else:
        print('Bạn nhận được thưởng 2000USD')
else:
    print('Bạn chưa đạt tiêu chí nhận thưởng')
```

### Thực hành:
Viết chương trình nhập vào kết quả 2 lần rút thăm may mắn. Nếu cả 2 lần đều không trúng thì in ra kết quả "Chúc bạn may mắn lần sau". Nếu cả 2 lần đều trúng thì in ra "Bingo!!!"

```python
lan_1=input('Nhập kết quả lần 1: ')
lan_2=input('Nhập kết quả lần 2: ')

if lan_1 == 'True':
    if lan_2 == 'True':
        print('Bingo')
    else:
        print('Chúc bạn may mắn lần sau')
else:
    print('Chúc bạn may mắn lần sau')
```

## Câu lệnh if…elif…else

```python
# Viết chương trình nhập vào số ngày trong tuần (2,3,4,5,6,7,8) và in ra từ tiếng anh tương ứng với ngày đó

week_day=int(input('Nhập vào số ngày: '))
if week_day==2:
    print('Monday')
elif week_day==3:
    print('Tuesday')
elif week_day==4:
    print('Wednesday')
elif week_day==5:
    print('Thursday')
elif week_day==6:
    print('Friday')
elif week_day==7:
    print('Saturday')
else:
    print('Sunday')
```

Note: tháng 2 28 ngày

```python
month_day=int(input('Nhập vào số tháng: '))
month_31=[1,3,5,7,8,10,12]
month_30=[4,6,9,11]
month_28=[2]

if month_day==2 or month_day==3 or month_day==5 or month_day==7 or month_day==8 or month_day==10 or month_day==12:
    print(f'Tháng {month_day} có 31 ngày')
elif month_day==4 or month_day==6 or month_day==9 or month_day==11:
    print(f'Tháng {month_day} có 30 ngày')
else:
    print(f'Tháng {month_day} có 28 ngày')
```
