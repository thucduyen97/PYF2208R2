### Bài 7: Một công ty vận tải có cách tính chi phí giao hàng như sau:
1. 10kilogram đầu tiên: \$5/kg
2. Từ kilogram thứ 11 đến 20: \$7/kg
3. Từ kilogram thứ 20 trở lên: \$10/kg

Viết chương trình nhập vào số kg cần giao(số nguyên) và in ra chi phí giao hàng tương ứng



````python
weight=int(input('Input the weight: '))

level1=5*weight
level2=5*10+7*(weight-10)
level3=5*10+7*10+10*(weight-20)
             
if 0<=weight<=10:
    print (f'Delivery cost la: {level1}')
elif 10<=weight<=20:
    print (f'Delivery cost la: {level2}')
else:
    print  (f'Delivery cost la: {level3}')
````

### Bài 10:
Tại một cửa hàng nọ có menu như sau:

1. Gà rán giá 35k
2. Hamburger giá 50k
3. Pesi giá 10k

Nhập số lượng món hàng cần mua

Tính và in ra tổng giá tiền

Nếu giá trị đơn hàng trên 2 triệu, giảm 10% cho khách, trên 4 triệu, giảm 15%

Tính và in ra tổng giá tiền được giảm

Tính và in ra tổng giá tiền khách cần trả

In ra dòng chữ "Cảm ơn quý khách đã mua hàng"



````python
"""Ví dụ:
Nhập số lượng gà rán khách mua:100
Nhập số lượng hamburger khách mua:100
Nhập số lượng pesi khách mua:15
Tổng giá tiền đơn hàng là: 8.650.000
Tổng tiền được giảm là: 1.297.500
Tổng tiền phải thanh toán: 7.352.500
Cảm ơn quý khách đã mua
"""
````

````python
#CHUA TINH TONG DISCOUNT DUOC

chicken=int(input('Input the quantity of chicken: '))
ham=int(input('Input the quantity of hamburger: '))
pepsi=int(input('Input the quantity of pepsi: '))
total=chicken*35+ham*50+pepsi*10
print('Total Price: ',total)

if total>2000:
    if total>4000:
        total%=15
        print('Total Discount: ',total)
        print('Total payment: ',total)
    else:
        total%=10
        print('Total Discount: ',total)
        print('Total payment: ',total)
else:
    print('Total Discount: 0')
    print('Total payment: ',total)

print('Thank you for purchasing')
````

### Bài 16: Viết một chương trình thực hiện phép tính đơn giản như sau:
```
    Nhập số thứ nhất: 16
    Nhập số thứ hai: 9
    Nhập toán tử (+, -, *, /): +
    Kết quả: 16 + 9 = 25
```

````python
num1=int(input('Input number 1: '))
num2=int(input('Input number 2: '))
math=input('Input math: ')

tong=num1+num2
hieu=num1-num2
tich=num1*num2
thuong=num1/num2

if math=='+':
    print(f'Tong 2 so la: {tong}')
elif math=='-':
    print(f'Hieu 2 so la: {hieu}')
elif math=='*':
    print(f'Tich 2 so la: {tich}')
else:
    print(f'Thuong 2 so la: {thuong}')
````

### Bài 17: Viết chương trình nhập vào 3 số nguyên và in ra màn hình theo thứ tự từ lớn đến bé.

````python
num1=int(input('Input number 1: '))
num2=int(input('Input number 2: '))
num3=int(input('Input number 3: '))

if num1>num2:
    if num1>num3:
        if num2>num3:
            print(f'The number order: {num1}, {num2}, {num3}')
        else:
            print(f'The number order: {num1}, {num3}, {num2}')
    else:
        print(f'The number order: {num3}, {num1}, {num2}')
else:
    if num1>num3:
        if num2>num3:
            print(f'The number order: {num2}, {num1}, {num3}')
        else:
            print(f'The number order: {num2}, {num3}, {num1}')
    else:
        print(f'The number order: {num3}, {num2}, {num1}')
````
