##

Trong Python, cơ bản có hai phạm vi hoạt động chủ yếu của một biến: Biến toàn cục (global), biến cục bộ (local)
 
Biến toàn cục (Global Variable) là biến được khai báo ngoài phạm vi của hàm 
=> có thể được sử dụng mọi nơi, 
=> có thể được truy cập bên trong hoặc bên ngoài hàm

```python
#Ví dụ: Biến toàn cục

x="Hàm và Module 1"
 
def foo():
    print("Trong hàm:", x)
foo()
print("Ngoài hàm:", x)
```

Khi sử dụng biến toàn cục bên trong một hàm thì bạn KHÔNG ĐƯỢC THAY ĐỔI GIÁ TRỊ CHO BIẾN. Như ví dụ dưới đây là sai, bởi vì mình đã thay đổi giá trị cho biến x.

```python
#Ví dụ: Thay đổi giá trị biến toàn cục

x = 10
 
def foo():
    x = x * 2
    print(x)
foo()
```

##
Biến cục bộ (Local Variable): biến được khai báo trong phạm vi của hàm.

```python
#Ví dụ: Biến cục bộ

def vidu(): 
    y = "Biến cục bộ" 
    print(y) vidu()
```

##
Từ khóa global: được sử dụng để khai báo một biến cục bộ trong hàm, giúp Python tự hiểu biến đó sẽ trỏ tới một biến cục bộ nằm bên ngoài hàm.

```python
#Ví dụ: Từ khoá Glogal

x = "global"
 
def foo():
    global x
    y = "local"
    x = x * 2
    print(x)
    print(y)
     
foo()
```

## GIÁ TRỊ TRẢ VỀ VÀ GỌI HÀM

Giới thiệu về lệnh return:
Hàm có thể có giá trị trả về hoặc không có giá trị trả về. Khi hàm có giá trị trả về, lệnh return được dùng để trả về một giá trị hoặc nhiều giá trị. Khi lệnh return được thực thi, hàm sẽ kết thúc và lệnh return là lệnh không bắt buộc có trong định nghĩa hàm.
Lệnh return có cú pháp như sau:

return [object] 

Trong đó: object là một đối tượng bất kì của một lớp nào đó, có thể là số (number), chuỗi (string), list, tuple, hàm (sẽ biết rõ hơn khi tìm hiểu decorator), lớp (class) hoặc thậm chí là bỏ trống – trường hợp bỏ trống thì object return về được tính là None

```python
#Ví dụ lệnh return

def calculator(a, b):
    sum = a + b
    return sum
print(sum)
```

```python
#Ví dụ lệnh return 2

def square(x):
    y = x*x
print(square(4))
#Output 
None
```

Khi chạy lệnh "print square(4)", chương trình sẽ in ra kết quả trả về của hàm square, nhưng do chúng ta không khai báo giá trị trả về trong hàm nên kết quả in ra là “None”.

```python
#Ví dụ

def square(x):
    return  x*x
print(square(4))

#Khi  gọi hàm “return” và thực thi mã nguồn, chương trình sẽ trả về giá trị “16”.
```

Hàm trả về nhiều giá trị:
Có thể sử dụng một lệnh return để trả về nhiều giá trị từ một hàm

```python
#Ví dụ

def calculator(a, b):
    summ = a + b
    sub = a - b
    mul = a * b
    return summ, sub, mul
print(summ, sub, mul)
```
