```python
class Personel:
    def __init__(self, idcode, name, age, startyear, department, level, salary):
        self.idcode = idcode
        self.name = name
        self.age = age
        self.startyear = startyear
        self.department = department
        self.level = level
        self.salary = salary
```

```python
from tkinter import *
from tkinter import messagebox
from tkinter import ttk
from PIL import Image, ImageTk

personel = Tk ()
img = Image.open ('/Users/thucduyen97/Documents/DA/Artboard 1.png')
photo = ImageTk.PhotoImage(img)
lab = Label(image = photo).place(x = 0, y = 0)
personel.title ('HR management system')
personel.geometry("1400x700")
bg = PhotoImage(file = '/Users/thucduyen97/Documents/DA/Artboard 1.png')

#create list to save list of ingradient in the company
personel_list = []

department_list = [
'Not choose',
'Human Resources',
'IT',
'Marketing',
'Business Analytics',
'Accounting'
]

level_list = [
'Not choose',
'1',
'2',
'3'
]

department_variable = StringVar(personel)
department_variable.set(department_list[0])

level_variable = StringVar(personel)
level_variable.set(level_list[0])

#create label
title_lbl = Label(personel, font = ('Roboto',14), bg = '#C6947B', fg = '#FFFFFF', text = 'Human Resources Management')
idcode_lbl = Label(personel, font = ('Roboto',14), bg = '#C6947B', fg = '#FFFFFF', text = 'Staff ID:')
name_lbl = Label(personel, font = ('Roboto',14), bg = '#C6947B', fg = '#FFFFFF', text = 'Name:')
age_lbl = Label(personel, font = ('Roboto',14), bg = '#C6947B', fg = '#FFFFFF', text = 'Age:')
startyear_lbl = Label(personel, font = ('Roboto',14), bg = '#C6947B', fg = '#FFFFFF', text = 'Start Year: ')
department_lbl = Label(personel, font = ('Roboto',14), bg = '#C6947B', fg = '#FFFFFF', text = 'Department: ')
level_lbl = Label(personel, font = ('Roboto',14), bg = '#C6947B', fg = '#FFFFFF', text = 'Level: ')
salary_lbl = Label(personel, font = ('Roboto',14), bg = '#C6947B', fg = '#FFFFFF', text = 'Salary: ')
search_lbl = Label(personel, font = ('Roboto',14), bg = '#C6947B', fg = '#FFFFFF', text = 'Search box: ')
note_lbl = Label(personel, font = ('Roboto',14), bg = '#C6947B', fg = '#FFFFFF', text = '*Note: Use staff ID/ staff name to search')

#create entry
idcode = Entry(personel, font = ('Roboto',12), bg = '#C6947B', fg = '#FFFFFF', width = 25)
name = Entry(personel, font = ('Roboto',12), bg = '#C6947B', fg = '#FFFFFF', width = 25)
age = Entry(personel, font = ('Roboto',12), bg = '#C6947B', fg = '#FFFFFF', width = 25)
startyear = Entry(personel, font = ('Roboto',12), bg = '#C6947B', fg = '#FFFFFF', width = 25)
department = OptionMenu(personel, department_variable, *department_list)
department.config(font = ('Roboto',12), fg = '#FFFFFF', width = 21)
level = OptionMenu(personel, level_variable, *level_list)
level.config(font = ('Roboto',12), fg = '#FFFFFF', width = 21)
salary = Entry(personel, font = ('Roboto',12), bg = '#C6947B', fg = '#FFFFFF', width = 25)
check = Entry(personel, font = ('Roboto',12), bg = '#C6947B', fg = '#FFFFFF', width = 25)

#function declaration
def add():
    personel_list.append(Personel(idcode.get(), name.get(), int(age.get()), int(startyear.get()),department_variable.get(),level_variable.get(),int(salary.get())))
    print([item.idcode for item in personel_list])
    print([item.name for item in personel_list])
    messagebox.showinfo('Personel',f'{name.get()} with Staff ID {idcode.get()} has been added succesfully to HR System')
    refresh_data(personel_list)
    
def reset():
    idcode.delete(0,150)
    name.delete(0,150)
    age.delete(0,150)
    startyear.delete(0,150)
    department_variable.set(department_list[0])
    level_variable.set(level_list[0])
    salary.delete(0,150)
    messagebox.showinfo('personel_list','Reset info')
    refresh_data(personel_list)
    
def search():
    search_idcode = check.get()
    search_name = check.get()
    exist = False 
    for item in personel_list:
        if item.idcode == search_idcode or item.name == search_name: 
            exist = True 
            messagebox.showinfo('Personel',f'{item.name} with staff ID: {item.idcode} is in HR system, information is as below: \n Age: {item.age}, start year: {item.startyear}, \n Department: {item.department}, level: {item.level}, salary: {item.salary}')
            break
    if exist == False:
        messagebox.showinfo('Personel',f'{check.get()} does not exist in list')
    refresh_data(personel_list)
    
def update():
    search_idcode = check.get()
    search_name = check.get()
    exist = False 
    for item in personel_list:
        if item.idcode == search_idcode or item.name == search_name: 
            exist = True
            item.name = name.get()
            item.age = age.get()
            item.startyear = startyear.get()
            item.department = department_variable.get()
            item.level = level_variable.get()
            item.salary = salary.get()
            messagebox.showinfo('Personel',f'{name.get()} with Staff ID {idcode.get()} has been updated succesfully')
            break
    if exist == False:
        messagebox.showinfo('Personel',f'{check.get()} does not exist in list')   
    print([item.idcode for item in personel_list])
    print([item.name for item in personel_list])
    refresh_data(personel_list)
    
def delete():
    delete_name = check.get()
    delete_idcode = check.get()
    exist = False
    for i in range(len(personel_list)):
        if personel_list[i].name == delete_name or personel_list[i].idcode == delete_idcode:
            exist = True
            del personel_list[i]
            messagebox.showinfo('Personel',f'{name.get()} with staff ID {idcode.get()} has been deleted')
            break
    if exist == False:
        messagebox.showinfo('Personel',f'{check.get()} does not exist in list')
    print([item.idcode for item in personel_list])
    print([item.name for item in personel_list])
    refresh_data(personel_list)
    
#create button
add_btn = Button(personel, font = ('Roboto',14), fg = '#000000', width = 5, text = 'Add', command = add)
reset_btn = Button(personel, font = ('Roboto',14), fg = '#000000', width = 5, text = 'Reset', command = reset)
search_btn = Button(personel, font = ('Roboto',14), fg = '#000000', width = 5, text = 'Search', command = search)
update_btn = Button(personel, font = ('Roboto',14), fg = '#000000', width = 5, text = 'Update', command = update)
delete_btn = Button(personel, font = ('Roboto',14), fg = '#000000', width = 5, text = 'Delete', command = delete)


#tree
def refresh_data(personel_list):
    for row in tree.get_children():
        tree.delete(row)
    for i in range(0, len(personel_list)):
        tree.insert('', END, values =(
            personel_list[i].idcode,
            personel_list[i].name,
            personel_list[i].age,
            personel_list[i].startyear,
            personel_list[i].department,
            personel_list[i].level,
            personel_list[i].salary
        ))
        
def item_selected(event):
    pass

columns = ('idcode', 'name', 'age','startyear','department','level','salary')
tree = ttk.Treeview(columns = columns, show = 'headings')

# define headings
tree.heading('idcode', text = 'ID Code')
tree.heading('name', text = 'Name')
tree.heading('age', text = 'Age')
tree.heading('startyear', text = 'Start year')
tree.heading('department', text = 'Department')
tree.heading('level', text = 'Level')
tree.heading('salary', text = 'Salary')

tree.bind('<<TreeviewSelect>>', item_selected)
tree.place(x = 5, y = 510)
refresh_data(personel_list)

#grid
title_lbl.place(x = 55, y = 8)
idcode_lbl.place(x = 5, y = 40)
name_lbl.place(x = 5, y = 70)
age_lbl.place(x = 5, y = 100)
startyear_lbl.place(x = 5, y = 130)
department_lbl.place(x = 5, y = 160)
level_lbl.place(x = 5, y = 190)
salary_lbl.place(x = 5, y = 220)

idcode.place(x = 95, y = 40)
name.place(x = 95, y = 70)
age.place(x = 95, y = 100)
startyear.place(x = 95, y = 130)
department.place(x = 95, y = 160)
level.place(x = 95, y = 190)
salary.place(x = 95, y = 220)

add_btn.place(x = 105, y = 260)
reset_btn.place(x = 205, y = 260)

search_lbl.place(x = 5, y = 320)
check.place(x = 95, y = 320)
note_lbl.place(x = 5, y = 345)
search_btn.place(x = 5, y = 385)
update_btn.place(x = 105, y = 385)
delete_btn.place(x = 205, y = 385)

personel.mainloop ()
```
